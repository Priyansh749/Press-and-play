```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import random

# Initialize variables for guess the number
secret_number = random.randint(1, 100)
coin_attempts = 0
number_result = ""

# Initial variables for rock, paper, scissors
player_choice = ""
computer_choice = ""
rps_result = ""

# Initial variables for the coin guessing game
coin_choice = random.choice(["Heads", "Tails"])
coin_attempts = 0
coin_result = ""
coin_guess = ""

# Guess the number
def gtn():
    global number_result
    def draw(canvas):
        canvas.draw_text("Guess the Number", (20, 40), 24, "White")
        canvas.draw_text("Enter your guess:", (20, 80), 18, "White")
        canvas.draw_text(number_result, (20, 120), 22, "red")

    # Handler for input field
    def input_handler(text):
        global number_result,coin_attempts
        guess = int(text)
        coin_attempts += 1
        if guess > 100 or guess <1:
            number_result = "Invalid input try again!"
        elif guess < secret_number:
                number_result = "Too low! Try again."
        elif guess > secret_number:
                number_result = "Too high! Try again."
        
        else:
            number_result = "Congratulations! You've guessed the number in " + str(coin_attempts) + " attempts."

    # Create a frame
    frame = simplegui.create_frame("Guess the Number", 800, 500, 100)
    frame.set_draw_handler(draw)

    # Add input field
    number_result = ""
    frame.add_input("Enter a guess:", input_handler, 100)

    # Start the frame
    frame.start()
    
# Rock, Paper, Scissors    

def rps():
    # Handler for player's choice buttons
    def rock_handler():
        global player_choice, computer_choice, rps_result
        player_choice = "Rock"
        computer_choice = random.choice(["Rock", "Paper", "Scissors"])
        determine_winner()

    def paper_handler():
        global player_choice, computer_choice, rps_result
        player_choice = "Paper"
        computer_choice = random.choice(["Rock", "Paper", "Scissors"])
        determine_winner()

    def scissors_handler():
        global player_choice, computer_choice, rps_result
        player_choice = "Scissors"
        computer_choice = random.choice(["Rock", "Paper", "Scissors"])
        determine_winner()

    # Determine the winner
    def determine_winner():
        global rps_result
        if player_choice == computer_choice:
            rps_result = "It's a tie!"
        elif (player_choice == "Rock" and computer_choice == "Scissors") or \
             (player_choice == "Paper" and computer_choice == "Rock") or \
             (player_choice == "Scissors" and computer_choice == "Paper"):
            rps_result = "User wins!"
        else:
            rps_result = "Computer wins!"

    # Handler to draw on the canvas
    def draw(canvas):
        canvas.draw_text("Rock, Paper, Scissors", (50, 50), 24, "White")
        canvas.draw_text("Player: " + player_choice, (50, 100), 18, "White")
        canvas.draw_text("Computer: " + computer_choice, (50, 130), 18, "White")
        canvas.draw_text(rps_result, (50, 160), 18, "White")

    # Create a frame
    frame = simplegui.create_frame("Rock, Paper, Scissors", 800, 500, 100)

    # Create buttons for player's choice
    frame.add_button("Rock", rock_handler)
    frame.add_button("Paper", paper_handler)
    frame.add_button("Scissors", scissors_handler)

    # Set the frame's draw handler
    frame.set_draw_handler(draw)

    # Start the frame
    frame.start()

    
def cgg():
    global coin_result, coin_choice, coin_attempts, coin_guess

    def draw(canvas):
        canvas.draw_text("Coin Guessing Game", (20, 40), 24, "White")
        canvas.draw_text(coin_result, (20, 120), 22, "red")

    # Handler for the "Heads" button
    def choose_heads():
        global coin_guess, coin_choice, coin_attempts, coin_result
        coin_guess = "Heads"
        check_guess()

    # Handler for the "Tails" button
    def choose_tails():
        global coin_guess, coin_choice, coin_attempts, coin_result
        coin_guess = "Tails"
        check_guess()

    # Function to check the user's guess
    def check_guess():
        global coin_result, coin_attempts, coin_choice, coin_guess

        coin_attempts += 1
        if coin_guess == coin_choice:
            coin_result = "Correct! You guessed " + coin_choice + " in " + str(coin_attempts) + " attempts."
            coin_attempts = 0
            new_game() # Start a new game automatically
        else:
            coin_result = "Incorrect guess it was " + coin_choice + " try again."
            new_game()

    # Function to start a new game
    def new_game():
        global coin_choice, coin_attempts, coin_result, coin_guess
        coin_choice = random.choice(["Heads", "Tails"])
        coin_guess = ""

    # Create a frame
    frame = simplegui.create_frame("Coin Guessing Game", 800, 500, 100)
    frame.set_draw_handler(draw)

    # Add buttons for "Heads" and "Tails"
    frame.add_button("Heads", choose_heads, 100)
    frame.add_button("Tails", choose_tails, 100)

    # Start the frame
    frame.start()
game_choice = ''
while game_choice != 'exit':
    game_choice = input('''
    Enter 1 to play Guess the number
    Enter 2 to play Rock, Paper, Scissors
    Enter 3 to play Coin guessing game
    Enter your choice: ''')

    str(game_choice)

    if game_choice == '1':
        gtn()
        game_choice = ''
    elif game_choice == '2':
        rps()
        game_choice = ''
    elif game_choice == '3':
        cgg()
        game_choice = ''
    elif game_choice == 'exit':
        game_choice = ''
        break
    else:
        print("Invalid input Enter 1, 2 or 3 only")
        game_choice = ''
```
